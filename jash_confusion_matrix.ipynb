```python
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
import seaborn as sns
```

```python
# true labels
jash_labels = ['service', 'product', 'policy', 'p&c', 'policy', 'resolution', 'p&c', 'service', 'policy', 'service', 'service', 'product', 'p&c', 'resolution', 'service', 'service']

# vedant labels
vedant_labels = ['service', 'product', 'policy', 'p&c', 'policy', 'resolution', 'p&c', 'service', 'policy', 'service', 'product', 'product', 'p&c', 'resolution', 'service', 'service']
```

```python
cm = confusion_matrix(jash_labels, vedant_labels, labels=['service', 'product', 'p&c', 'policy', 'resolution'])
```

```python
plt.figure(figsize=(10,7))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=['service', 'product', 'p&c', 'policy', 'resolution'], yticklabels=['service', 'product', 'p&c', 'policy', 'resolution'])
plt.xlabel('Predicted (Vedant)')
plt.ylabel('True (Jash)')
plt.title('Confusion Matrix')
plt.show()
```
